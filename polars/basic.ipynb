# Learn Polars

```python
import polars as pl
import numpy as np
```

```python
np.random.seed(12)

df = pl.DataFrame(
    {
        "nrs": [1, 1, 2, 2, 3, 3],
        "random": np.random.rand(6),
    }
)

df
```

```python
df.sort("random")
```

```python
df.sort("random").groupby("nrs").last()
```
